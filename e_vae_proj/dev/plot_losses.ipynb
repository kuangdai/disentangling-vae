```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
hist_kl_rec = np.mean(np.load('hist_constr_kl_rec.npy'), axis=1)
hist_kl_kl = np.mean(np.load('hist_constr_kl_KL.npy'), axis=1)
hist_kl_loss = np.mean(np.load('hist_constr_kl_loss.npy'), axis=1)
hist_kl_lambda = np.mean(np.load('hist_constr_kl_lambda.npy'), axis=1)

hist_recon_rec = np.mean(np.load('hist_constr_recon_rec.npy'), axis=1)
hist_recon_kl = np.mean(np.load('hist_constr_recon_KL.npy'), axis=1)
hist_recon_loss = np.mean(np.load('hist_constr_recon_loss.npy'), axis=1)
hist_recon_lambda = np.mean(np.load('hist_constr_recon_lambda.npy'), axis=1)
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2, dpi=200, facecolor='w')

# recon
ax = axes[0, 0]
ax.plot(hist_kl_rec, linewidth=.8)
ax.tick_params('both', length=2, pad=0.5, which='major', labelsize=5)
ax.set_ylabel('Reconstruction Error', fontsize=6)

# kl
ax = axes[0, 1]
ax.plot(hist_kl_kl, linewidth=.8)
ax.tick_params('both', length=2, pad=0.5, which='major', labelsize=5)
ax.set_ylabel('Total KL', fontsize=6)

# loss
ax = axes[1, 0]
ax.plot(hist_kl_loss, linewidth=.8)
ax.tick_params('both', length=2, pad=0.5, which='major', labelsize=5)
ax.set_xlabel('Epoch', fontsize=6)
ax.set_ylabel('Loss', fontsize=6)

# lambda
ax = axes[1, 1]
ax.plot(hist_kl_lambda, linewidth=.8)
ax.tick_params('both', length=2, pad=0.5, which='major', labelsize=5)
ax.set_xlabel('Epoch', fontsize=6)
ax.set_ylabel(r'$\lambda$, penalty factor', fontsize=6)

fig.suptitle(r'$\epsilon$-VAE for MNIST with $\epsilon_{KL} =$ 1.0, $LR_{\lambda} =$ 0.01, $\Delta L =$ 2', fontsize=8)
plt.savefig('evae_kl_mnist_eps1.0_LR.01_nd2.png', bbox_inches='tight', pad_inches=0.05)
plt.show()
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2, dpi=200, facecolor='w')

# recon
ax = axes[0, 0]
ax.plot(hist_recon_rec, linewidth=.8)
ax.tick_params('both', length=2, pad=0.5, which='major', labelsize=5)
ax.set_ylabel('Reconstruction Error', fontsize=6)

# kl
ax = axes[0, 1]
ax.plot(hist_recon_kl, linewidth=.8)
ax.tick_params('both', length=2, pad=0.5, which='major', labelsize=5)
ax.set_ylabel('Total KL', fontsize=6)

# loss
ax = axes[1, 0]
ax.plot(hist_recon_loss, linewidth=.8)
ax.tick_params('both', length=2, pad=0.5, which='major', labelsize=5)
ax.set_xlabel('Epoch', fontsize=6)
ax.set_ylabel('Loss', fontsize=6)

# lambda
ax = axes[1, 1]
ax.plot(hist_recon_lambda, linewidth=.8)
ax.tick_params('both', length=2, pad=0.5, which='major', labelsize=5)
ax.set_xlabel('Epoch', fontsize=6)
ax.set_ylabel(r'$\lambda$, penalty factor', fontsize=6)

fig.suptitle(r'$\epsilon$-VAE for MNIST with $\epsilon_{recon} =$ 180, $LR_{\lambda} =$ 0.01, $\Delta L =$ 2', fontsize=8)
plt.savefig('evae_recon_mnist_eps180_LR.01_nd2.png', bbox_inches='tight', pad_inches=0.05)
plt.show()
```
