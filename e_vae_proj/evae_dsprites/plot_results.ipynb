```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt
plt.rcParams['axes.linewidth'] = .5
plt.rcParams['xtick.major.width'] = .5
plt.rcParams['ytick.major.width'] = .5
plt.rcParams['xtick.minor.width'] = .0
plt.rcParams['ytick.minor.width'] = .0
plt.rcParams['axes.labelpad'] = 1
plt.rcParams['axes.titlepad'] = 3
plt.rcParams['font.size'] = 6
```

```python
# constraint
cons = 'rec'

# read grid
epses = np.loadtxt(f'grid_eps_{cons}')
nlats = np.loadtxt('grid_nlats').astype(int)

# read data
LCM = np.load(f'results_{cons}/metric_LCM.npy')
MIG = np.load(f'results_{cons}/metric_MIG.npy')
AAD = np.load(f'results_{cons}/metric_AAD.npy')
REC = np.load(f'results_{cons}/metric_REC.npy')
KL = np.load(f'results_{cons}/metric_KL.npy')
LOSS = np.load(f'results_{cons}/metric_LOSS.npy')

REC /= 4096
LOSS /= 4096
KL /= nlats

# some unconverged ones
MIG[MIG > MIG.max() * .9] = 0.
AAD[np.isnan(AAD)] = 0.
AAD[AAD > AAD.max() * .9] = 0.
```

```python
# settings
xticks_id = np.array([1, 4, 8, 13, 18, 23])
xticks = nlats[::-1][xticks_id]
xticks_id -= 1
yticks_id = [0, 3, 6, 9, 13, 18, 22, 27, 31]
yticks = epses[yticks_id]
cm_loss = 'gist_stern'
cm_metric = 'turbo'

fig, axes = plt.subplots(nrows=2, ncols=3, dpi=200, facecolor='w')
plt.subplots_adjust(wspace=-.2)


# loss
ax = axes[0, 0]
im = ax.imshow(LOSS[:, ::-1][:, 1:], origin='lower', cmap=cm_loss, vmin=0, vmax=0.14)
cbar = fig.colorbar(im, ax=ax, pad=.02)
cbar.ax.tick_params('both', length=2, pad=0.5, which='major', labelsize=5)
ax.set_xticks(xticks_id)
ax.set_yticks(yticks_id)
ax.set_xticklabels([])
ax.set_yticklabels(yticks, fontsize=5)
yl = ax.get_yticklabels()
yl[2] = plt.Text(0, 6, '1e-04')
ax.set_yticklabels(yl, fontsize=5)
ax.set_ylabel(r'Normalized $\epsilon$', fontsize=6)
ax.set_title(r'Loss of $\epsilon$-VAEs', fontsize=6)
ax.tick_params('both', length=2, width=.5, pad=0.5, which='major')
ax.set_xticks(np.arange(len(nlats) - 1) + .5, minor=True)
ax.set_yticks(np.arange(len(epses)) + .5, minor=True)
ax.grid(which='minor', c='gray', lw=.1)

# rec
ax = axes[0, 1]
im = ax.imshow(REC[:, ::-1][:, 1:], origin='lower', cmap=cm_loss, vmin=0, vmax=0.14)
cbar = fig.colorbar(im, ax=ax, pad=.02)
cbar.ax.tick_params('both', length=2, pad=0.5, which='major', labelsize=5)
ax.set_xticks(xticks_id)
ax.set_yticks(yticks_id)
ax.set_xticklabels([])
ax.set_yticklabels([])
ax.set_title(r'Reconstr. loss of $\epsilon$-VAEs', fontsize=6)
ax.tick_params('both', length=2, width=.5, pad=0.5, which='major')
ax.set_xticks(np.arange(len(nlats) - 1) + .5, minor=True)
ax.set_yticks(np.arange(len(epses)) + .5, minor=True)
ax.grid(which='minor', c='gray', lw=.1)

# KL
ax = axes[0, 2]
im = ax.imshow(KL[:, ::-1][:, 1:], origin='lower', cmap=cm_loss, vmin=0, vmax=8)
cbar = fig.colorbar(im, ax=ax, pad=.02)
cbar.ax.tick_params('both', length=2, pad=0.5, which='major', labelsize=5)
ax.set_xticks(xticks_id)
ax.set_yticks(yticks_id)
ax.set_xticklabels([])
ax.set_yticklabels([])
ax.set_title(r'KL loss of $\epsilon$-VAEs', fontsize=6)
ax.tick_params('both', length=2, width=.5, pad=0.5, which='major')
ax.set_xticks(np.arange(len(nlats) - 1) + .5, minor=True)
ax.set_yticks(np.arange(len(epses)) + .5, minor=True)
ax.grid(which='minor', c='gray', lw=.1)

# LCM
ax = axes[1, 0]
im = ax.imshow(LCM[:, ::-1][:, 1:], origin='lower', cmap=cm_metric, vmin=0.2, vmax=1.)
cbar = fig.colorbar(im, ax=ax, pad=.02)
cbar.ax.tick_params('both', length=2, pad=0.5, which='major', labelsize=5)
ax.set_xticks(xticks_id)
ax.set_yticks(yticks_id)
ax.set_xticklabels(xticks, fontsize=5)
ax.set_yticklabels(yticks, fontsize=5)
yl = ax.get_yticklabels()
yl[2] = plt.Text(0, 6, '1e-04')
ax.set_yticklabels(yl, fontsize=5)
ax.set_xlabel(r'Latent size', fontsize=6)
ax.set_ylabel(r'Normalized $\epsilon$', fontsize=6)
ax.set_title(r'LCM of $\epsilon$-VAEs', fontsize=6)
ax.tick_params('both', length=2, width=.5, pad=0.5, which='major')
ax.set_xticks(np.arange(len(nlats) - 1) + .5, minor=True)
ax.set_yticks(np.arange(len(epses)) + .5, minor=True)
ax.grid(which='minor', c='gray', lw=.1)

# MIG
ax = axes[1, 1]
im = ax.imshow(MIG[:, ::-1][:, 1:], origin='lower', cmap=cm_metric, vmin=0, vmax=.005)
cbar = fig.colorbar(im, ax=ax, pad=.02)
cbar.ax.tick_params('both', length=2, pad=0.5, which='major', labelsize=5)
ax.set_xticks(xticks_id)
ax.set_yticks(yticks_id)
ax.set_xticklabels(xticks, fontsize=5)
ax.set_yticklabels([])
ax.set_xlabel(r'Latent size', fontsize=6)
ax.set_title(r'MIG of $\epsilon$-VAEs', fontsize=6)
ax.tick_params('both', length=2, width=.5, pad=0.5, which='major')
ax.set_xticks(np.arange(len(nlats) - 1) + .5, minor=True)
ax.set_yticks(np.arange(len(epses)) + .5, minor=True)
ax.grid(which='minor', c='gray', lw=.1)

# AAD
ax = axes[1, 2]
im = ax.imshow(AAD[:, ::-1][:, 1:], origin='lower', cmap=cm_metric, vmin=0, vmax=.005)
cbar = fig.colorbar(im, ax=ax, pad=.02)
cbar.ax.tick_params('both', length=2, pad=0.5, which='major', labelsize=5)
ax.set_xticks(xticks_id)
ax.set_yticks(yticks_id)
ax.set_xticklabels(xticks, fontsize=5)
ax.set_yticklabels([])
ax.set_xlabel(r'Latent size', fontsize=6)
ax.set_title(r'MID of $\epsilon$-VAEs', fontsize=6)
ax.tick_params('both', length=2, width=.5, pad=0.5, which='major')
ax.set_xticks(np.arange(len(nlats) - 1) + .5, minor=True)
ax.set_yticks(np.arange(len(epses)) + .5, minor=True)
ax.grid(which='minor', c='gray', lw=.1)

# save
plt.savefig(f'results_{cons}/metrics.pdf', bbox_inches='tight', pad_inches=0.05)
plt.show()
```
