```python
%pylab inline
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

plt.rcParams['axes.linewidth'] = .5
plt.rcParams['xtick.major.width'] = .5
plt.rcParams['ytick.major.width'] = .5
plt.rcParams['xtick.minor.width'] = .0
plt.rcParams['ytick.minor.width'] = .0
plt.rcParams['axes.labelpad'] = 1
plt.rcParams['axes.titlepad'] = 3
plt.rcParams['font.size'] = 6
```

```python
from pathlib import Path
import os

# absolute path
my_path = Path(os.path.abspath('')).parent.resolve().expanduser()
main_path = my_path.parent.parent

# get posterior traversals for each model
bvae_posterior_traversals = mpimg.imread(main_path / 'results' / 'bvaeH_mnist_100ep' / 'z10_b4.0_s0' / 'posterior_traversals.png')
btcvae_posterior_traversals = mpimg.imread(main_path / 'results' / 'btcvae_mnist_100ep' / 'z10_a1.0_b6.0_g1.0_s0' / 'posterior_traversals.png')
evae_kl_posterior_traversals = mpimg.imread(main_path / 'results' / 'evae_mnist' / 'kl_100ep_z10_e0.1_s0' / 'posterior_traversals.png')
evae_rec_posterior_traversals = mpimg.imread(main_path / 'results' / 'evae_mnist' / 'rec_100ep_z10_e0.2_s0' / 'posterior_traversals.png')

# get prior traversals for each model
bvae_prior_traversals = mpimg.imread(main_path / 'results' / 'bvaeH_mnist_100ep' / 'z10_b4.0_s0' / 'prior_traversals.png')
btcvae_prior_traversals = mpimg.imread(main_path / 'results' / 'btcvae_mnist_100ep' / 'z10_a1.0_b6.0_g1.0_s0' / 'prior_traversals.png')
evae_kl_prior_traversals = mpimg.imread(main_path / 'results' / 'evae_mnist' / 'kl_100ep_z10_e0.1_s0' / 'prior_traversals.png')
evae_rec_prior_traversals = mpimg.imread(main_path / 'results' / 'evae_mnist' / 'rec_100ep_z10_e0.2_s0' / 'prior_traversals.png')

# get reconstructions & traversal images for each model
bvae_rec_traverse = mpimg.imread(main_path / 'results' / 'bvaeH_mnist_100ep' / 'z10_b4.0_s0' / 'reconstruct_traverse.png')
btcvae_rec_traverse = mpimg.imread(main_path / 'results' / 'btcvae_mnist_100ep' / 'z10_a1.0_b6.0_g1.0_s0' / 'reconstruct_traverse.png')
evae_kl_rec_traverse = mpimg.imread(main_path / 'results' / 'evae_mnist' / 'kl_100ep_z10_e0.1_s0' / 'reconstruct_traverse.png')
evae_rec_rec_traverse = mpimg.imread(main_path / 'results' / 'evae_mnist' / 'rec_100ep_z10_e0.2_s0' / 'reconstruct_traverse.png')
```

```python
fig, axs = plt.subplots(nrows=2, ncols=2, dpi=200, facecolor='w', tight_layout=True)
plt.subplots_adjust(wspace=-.2)

ax = axs[0,0]
ax.imshow(bvae_posterior_traversals)
ax.axis('off')
ax.set_title(r'$\beta_{H}$-VAE, $\beta = $4')

ax = axs[0,1]
ax.imshow(btcvae_posterior_traversals)
ax.axis('off')
ax.set_title(r'$\beta$-TCVAE, $\alpha = \gamma = $1, $\beta = $6')

ax = axs[1,0]
ax.imshow(evae_kl_posterior_traversals)
ax.axis('off')
ax.set_title(r'$\epsilon$-VAE, $\epsilon_{KL} = $0.1')

ax = axs[1,1]
ax.imshow(evae_rec_posterior_traversals)
ax.axis('off')
ax.set_title(r'$\epsilon$-VAE, $\epsilon_{rec} = $0.1')

fig.suptitle('Posterior traversals')
plt.savefig('posterior-travers-compare.png', tight_layout=True)
```
